```python
import pandas as pd
import folium
```

```python
addresses = pd.read_csv("places.csv", delimiter=";")
```

```python
locations = addresses[['latitude', 'longitude']]
locationlist = locations.values.tolist()

map = folium.Map(location=[40.4686, 19.48318],zoom_start=3)
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point], popup=addresses['street'][point]).add_to(map)
```
